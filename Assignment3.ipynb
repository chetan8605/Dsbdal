# Descriptive Statistics for Titanic Dataset

```python
ti = df=pd.read_csv("Titanic.csv")
```

```python
ti.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived', data=ti)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived', hue='Sex', data=ti, palette='RdBu_r')
```

```python
ti.groupby('Sex')["Survived"].agg(['sum'])
```

```python
ti.groupby('Sex')["Survived"].agg(['std'])
```

```python
ti.groupby('Sex')["Survived"].agg(['mean'])
```

```python
ti.groupby('Sex')["Survived"].describe()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived', hue='Pclass', data=ti, palette='rainbow')
```

```python
ti.groupby("Pclass")['Survived'].agg(["sum"])
```
