```python
less = sales[sales['Fuel_efficiency']>=19]
```

```python
less = less[less['Fuel_efficiency']<25]
less
```

```python
middle = sales[sales['Fuel_efficiency']>=25]
middle = middle[middle['Fuel_efficiency']<30]
middle
```

```python
more = sales[sales['Fuel_efficiency']>=30]
more
```

```python
l_mean_sal = less['Fuel_efficiency'].mean()
l_median_sal = less['Fuel_efficiency'].median()
l_max_sal = less['Fuel_efficiency'].max()
l_min_sal = less['Fuel_efficiency'].min()
l_std_sal = less['Fuel_efficiency'].std()
print('Mean Fuel_efficiency of l group ',l_mean_sal)
print('Median Fuel_efficiency of l group ',l_median_sal)
print('Maximum Fuel_efficiency of l group ',l_max_sal)
print('Minimum Fuel_efficiency of l group ',l_min_sal)
print('Standard deviation of Fuel_efficiency of l group ',l_max_sal)
```

```python
mid_mean_sal = middle['Fuel_efficiency'].mean()
mid_median_sal = middle['Fuel_efficiency'].median()
mid_max_sal = middle['Fuel_efficiency'].max()
mid_min_sal = middle['Fuel_efficiency'].min()
mid_std_sal = middle['Fuel_efficiency'].std()
print('Mean Fuel_efficiency of mid group ',mid_mean_sal)
print('Median Fuel_efficiency of mid group ',mid_median_sal)
print('Maximum Fuel_efficiency of mid group ',mid_max_sal)
print('Minimum Fuel_efficiency of mid group ',mid_min_sal)
print('Standard deviation of Fuel_efficiency of mid group ',mid_max_sal)
```

```python
m_mean_sal = more['Fuel_efficiency'].mean()
m_median_sal = more['Fuel_efficiency'].median()
m_max_sal = more['Fuel_efficiency'].max()
m_min_sal = more['Fuel_efficiency'].min()
m_std_sal = more['Fuel_efficiency'].std()
print('Mean Fuel_efficiency of m group ',m_mean_sal)
print('Median Fuel_efficiency of m group ',m_median_sal)
print('Maximum Fuel_efficiency of m group ',m_max_sal)
print('Minimum Fuel_efficiency of m group ',m_min_sal)
print('Standard deviation of Fuel_efficiency of m group ',m_max_sal)
```
